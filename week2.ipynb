# DATA ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ds = pd.read_csv("/home/nbuser/library/assignment 1/googleplaystore.csv")
ds['Installs']=ds['Installs'].apply(lambda a:a.split('+')[0])  
se=ds['Installs'].apply(lambda a:a.split(','))              

def add_list(x):
    sum=' '
    for i in range(0,len(x)):
        sum+=x[i]
    return int(sum)  

ds['Installs']=se.apply(lambda a:add_list(a)) 

ds['Type'].value_counts()

plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
sns.countplot(x='Type',data=ds)
plt.title("Number of Apps Available: Free v/s Paid")


plt.subplot(1,2,2)
sns.barplot(x='Type',y='Installs',data=ds,ci=None)
plt.title("Most installed: Free v/s Paid")
plt.tight_layout()
```

The first graph comapares the number of apps that are free and paid. The second graph shows the most installed apps are Free to install.

```python
#PRICE DISTRIBUTION PIE CHART
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds = pd.read_csv("/home/nbuser/library/assignment 1/googleplaystore.csv")


newPrice = []

for row in ds.Price:
    if row!= "0":
        newrow = float(row[1:])
    else:
        newrow = 0 
        
    newPrice.append(newrow)
        
ds.Price = newPrice

plt.pie(ds.Type.value_counts(), labels=['Free', 'Paid'], autopct='%1.1f%%')
```

the above pie chart shows that 92.6% of the Apps are Free in the Play Store
